```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from bokeh.plotting import figure, show, output_file
```

```python
my_file = h5py.File("alaska_solar_data_storage.hdf5", 'r')
```

```python
fairbanks = my_file.get('Fairbanks')
```

```python
def hdf5_to_dataframe(hdf5_filename, location_name, panel_name):
    """This function extracts data from an HDF5 file and loads it into a pandas dataframe"""
    #Load the HDF5 file data
    hdf5_file = h5py.File('{}.hdf5'.format(hdf5_filename), 'r')
    hdf5_location = hdf5_file.get(location_name)
    panel_location = hdf5_location.get(panel_name)
    dataframe = pd.DataFrame()
    for keys in panel_location.keys():
        dataframe[str(keys)] = panel_location[str(keys)]
    return dataframe
```

```python
result=pd.DataFrame(columns=['Month', 'Year'])
a=0
for name in fairbanks:
    base= hdf5_to_dataframe('alaska_solar_data_storage','Fairbanks', name )
    result = pd.merge(result, base, on = ['Month', 'Year'], how='outer',suffixes=(a, a+1))
    a=a+1
```

```python
result['Date'] = ""
for i in range(len(result)):
    result['Date'][i] = str(result['Year'][i]) + '-' + str(result['Month'][i])    
    # Change the date into a datetime format
    result['Date'] = pd.to_datetime(result['Date'])
    
result.drop(['Year','Month'],axis = 1,inplace = True)

result
```

```python
col=result['Date']
result= result.drop(['Date'], axis=1)
#result.loc["mean"]=result.apply(np.mean)
result
```

```python
result['Mean'] = result.mean(1)
result["STD"] = result.std(1)
result
```

```python

groups = np.linspace(0,102)
xs = groups
yerrs = result['STD']
ys = result['Mean']

output_file('errorbars.html')

# plot the points
p = figure(title='Fairbanks errorbars', width=800, height=400)

p.xaxis.axis_label = 'Month'
p.yaxis.axis_label = 'Production'


p.circle(xs, ys, color='red', size=5, line_alpha=0)


# create the coordinates for the errorbars
err_xs = []
err_ys = []

for x, y, yerr in zip(xs, ys, yerrs):
    err_xs.append((x, x))
    err_ys.append((y - yerr, y + yerr))
    
p.vbar(x=groups, width=0.5, bottom=0,
       top=result['Mean'], color="firebrick")
# plot them
p.multi_line(err_xs, err_ys, color='blue')

show(p)
```
