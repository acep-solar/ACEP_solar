```python
ARappoport = pd.read_excel('../../Cleaned_data/AnchorageSolar_cleaned/17ARappoport_cleaned.xlsx')
cwilcox = pd.read_excel('../../Cleaned_data/AnchorageSolar_cleaned/17cwilcox_cleaned.xlsx')
dgroth = pd.read_excel('../../Cleaned_data/AnchorageSolar_cleaned/17dgroth_cleaned.xlsx')
JFredston = pd.read_excel('../../Cleaned_data/AnchorageSolar_cleaned/17JFredston_cleaned.xlsx')
jreynold = pd.read_excel('../../Cleaned_data/AnchorageSolar_cleaned/17jreynold_cleaned.xlsx')
jrpatee = pd.read_excel('../../Cleaned_data/AnchorageSolar_cleaned/17jrpatee_cleaned.xlsx')
jslemons = pd.read_excel('../../Cleaned_data/AnchorageSolar_cleaned/17jslemons_cleaned.xlsx')
MPennington = pd.read_excel('../../Cleaned_data/AnchorageSolar_cleaned/17MPennington_cleaned.xlsx')
NPeters = pd.read_excel('../../Cleaned_data/AnchorageSolar_cleaned/17NPeters_cleaned.xlsx')
pjohnsony = pd.read_excel('../../Cleaned_data/AnchorageSolar_cleaned/17pjohnson_cleaned.xlsx')
rrogers = pd.read_excel('../../Cleaned_data/AnchorageSolar_cleaned/17rrogers.xlsx')
SherKim = pd.read_excel('../../Cleaned_data/AnchorageSolar_cleaned/17SherKim_cleaned.xlsx')
```

```python
ARappoport_new = add_time_col(ARappoport)
cwilcox_new = add_time_col(cwilcox)
dgroth_new = add_time_col(dgroth)
JFredston_new = add_time_col(JFredston)
jreynold_new = add_time_col(jreynold)
jrpatee_new = add_time_col(jrpatee)
jslemons_new = add_time_col(jslemons)
MPennington_new = add_time_col(MPennington)
NPeters_new = add_time_col(NPeters)
pjohnsony_new = add_time_col(pjohnsony)
rrogers_new = add_time_col(rrogers)
SherKim_new = add_time_col(SherKim)
```

```python
ARa_annual_values,ARa_month = annual_Norm(ARappoport_new)
cwi_annual_values,cwi_month = annual_Norm(cwilcox_new)
dgr_annual_values,dgr_month = annual_Norm(dgroth_new)
JFr_annual_values,JFr_month = annual_Norm(JFredston_new)
jre_annual_values,jre_month = annual_Norm(jreynold_new)
jrp_annual_values,jrp_month = annual_Norm(jrpatee_new)
jsl_annual_values,jsl_month = annual_Norm(jslemons_new)
MPe_annual_values,MPe_month = annual_Norm(MPennington_new)
NPe_annual_values,NPe_month = annual_Norm(NPeters_new)
pjo_annual_values,pjo_month = annual_Norm(pjohnsony_new)
rro_annual_values,rro_month = annual_Norm(rrogers_new)
She_annual_values,She_month = annual_Norm(SherKim_new)
```

```python
new_ARa = pd.DataFrame({'Date':ARa_month,'ARappoport':ARa_annual_values})
new_cwi = pd.DataFrame({'Date':cwi_month,'cwilcox':cwi_annual_values})
new_dgr = pd.DataFrame({'Date':dgr_month,'dgroth':dgr_annual_values})
new_JFr = pd.DataFrame({'Date':JFr_month,'JFredston':JFr_annual_values})
new_jre = pd.DataFrame({'Date':jre_month,'jreynold':jre_annual_values})
new_jrp = pd.DataFrame({'Date':jrp_month,'jrpatee':jrp_annual_values})
new_jsl = pd.DataFrame({'Date':jsl_month,'jslemons':jsl_annual_values})
new_MPe = pd.DataFrame({'Date':MPe_month,'MPennington':MPe_annual_values})
new_NPe = pd.DataFrame({'Date':NPe_month,'NPeters':NPe_annual_values})
new_pjo = pd.DataFrame({'Date':pjo_month,'pjohnsony':pjo_annual_values})
new_rro = pd.DataFrame({'Date':rro_month,'rrogers':rro_annual_values})
new_She = pd.DataFrame({'Date':She_month,'SherKim':She_annual_values})
result = pd.merge(new_ARa,new_cwi,on='Date',how='outer')
result = pd.merge(result,new_dgr,on='Date',how='outer')
result = pd.merge(result,new_JFr,on='Date',how='outer')
result = pd.merge(result,new_jre,on='Date',how='outer')
result = pd.merge(result,new_jrp,on='Date',how='outer')
result = pd.merge(result,new_jsl,on='Date',how='outer')
result = pd.merge(result,new_MPe,on='Date',how='outer')
result = pd.merge(result,new_NPe,on='Date',how='outer')
result = pd.merge(result,new_pjo,on='Date',how='outer')
result = pd.merge(result,new_rro,on='Date',how='outer')
Anchorage_17 = pd.merge(result,new_She,on='Date',how='outer')
Anchorage_17['Mean'] = Anchorage_17.mean(axis=1)
Anchorage_17 = Anchorage_17.sort_values(by = ['Date'])
Anchorage_17
```
