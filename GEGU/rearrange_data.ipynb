```python
def drop_the_hours (dataframe):
    """This function is to drop the hours in the first columns of the daily data
    
       Arg:
       dataframe(string):The name of the dataframe that you want to change
       
       Return:
       the dataframe after changed
    """
    
    date_time = []
    
    for i in range(len(dataframe)):
        date_time.append(dataframe['Date/Time'][i].rsplit(' ',2)[0])
    
    dataframe['Date/Time']=date_time
    dataframe.drop(dataframe.tail(1).index,inplace=True)
    
    return(dataframe)

def seperate_date(dataframe):
    """This function will create a new dataframe that contains the columns in order of 'Year','Month','Day','Energy'
    
       Args:
       dataframeï¼ˆstring):The name of the dataframe contain the data you want to use
       
       Return:
       A new dataframe
    """
    
    dataframe_new = pd.DataFrame(columns=['Year','Month','Day','Energy'])
    year=[]
    month=[]
    day=[]
    energy=[]
    
    for i in range(len(dataframe)):
        year.append(int(dataframe['Date/Time'][i].rsplit('-',2)[0]))
        month.append(int(dataframe['Date/Time'][i].rsplit('-',2)[1]))
        day.append(int(dataframe['Date/Time'][i].rsplit('-',2)[2]))
        energy.append(dataframe['Energy Produced (Wh)'][i]/1000)
    
    dataframe_new['Year']=year
    dataframe_new['Month']=month
    dataframe_new['Day']=day
    dataframe_new['Energy'] = energy
    
    return(dataframe_new)
```
