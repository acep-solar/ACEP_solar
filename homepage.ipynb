```python
import os
import folium
import pandas as pd
import branca
```

```python
def read_file(file_path):
    all_the_text = open(file_path).read()
    # print type(all_the_text)
    return all_the_text
```

```python
alaska= folium.Map(location = [63.977161, -152.024667], zoom_start = 4)
alaska
```

```python

#Anchorage
folium.Marker([61.193625, -149.694974], popup = 'Anchorage, AK.', 
              tooltip = 'Anchorage, AK.').add_to(alaska)
#Kodiak
folium.Marker([57.794934, -152.392466], popup = 'Kodiak, AK.', 
              tooltip = 'Kodiak, AK.').add_to(alaska)
#Nome
folium.Marker([64.516804, -165.399455], popup = 'Nome, AK.', 
              tooltip = 'Nome, AK.').add_to(alaska)
#Juneau
folium.Marker([58.306096, -134.422802], popup = 'Juneau, AK.', 
              tooltip = 'Juneau, AK.').add_to(alaska)
#Prudhoe Bay
folium.Marker([70.212820, -148.408886], popup = 'Prudhoe Bay, AK.', 
              tooltip = 'Prudhoe Bay, AK.').add_to(alaska)
#Kotzebue
folium.Marker([66.888948, -162.512280], popup = 'Kotzebue, AK.',
              tooltip = 'Kotzebue, AK.').add_to(alaska)
#Bethel
folium.Marker([60.794938, -161.770716], popup = 'Bethel, AK.', 
              tooltip = 'Bethel, AK.').add_to(alaska)
```

```python
#Fairbanks

popupfigure = read_file("popup_fairbanks.html")

html = """

    {0}
    
    
    <a href="https://acep-solar.github.io/ACEP_solar/webpages/fairbanks.html" title="Go ahead" target="_blank" name="girl">Click Here for More Infromation</a>
   

"""
html = html.format(popupfigure)

  
iframe = branca.element.IFrame(html = html, width=650, height=400)
popup = folium.Popup(iframe, max_width=850)

folium.Marker([64.838033, -147.668970], popup=popup,
              tooltip='Fairbanks, AK.').add_to(alaska)

alaska
```

```python
alaska.save(os.path.join( 'index.html'))
```
